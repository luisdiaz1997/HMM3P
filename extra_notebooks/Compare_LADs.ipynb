```python
hmmp_df = bf.read_table(hmmp_bed, schema='bed9')
lad1_df = bf.read_table(lad1_bed, schema='bed9')
lad1_df.name = 'T1-LAD'
lad1_df.rgb = '116,173,209'
# lad1_df = lad1_df.drop(columns=['score', 'strand', 'thickStart', 'thickEnd'])

lad2_df = bf.read_table(lad2_bed, schema='bed9')
lad2_df.name = 'T2-LAD'
lad2_df.rgb = '250,250,210'
# lad2_df  = lad2_df.drop(columns=['score', 'strand', 'thickStart', 'thickEnd'])
lad_df = pd.concat((lad1_df, lad2_df))
```

```python
lad_df.name.unique()
```

```python
def bin_track(chrombins, bedtrack, columns = ['chrom', 'start', 'end', 'name', 'rgb'], rgb=True, lad=False):
    binned_track = bf.overlap(df1=chrombins, df2=bedtrack.loc[:, columns], how='left')
    binned_track = binned_track.iloc[:, [0,1,2,6,7]]
    binned_track.columns= columns
    mask = (binned_track.name == '.') | (binned_track.name.isna())
    if lad:
        binned_track.loc[mask, 'name'] = 'non-LAD'
        binned_track.loc[mask, 'rgb'] = '244, 109, 67'
    else:
        
        binned_track.loc[mask, 'name'] = 'N'
        
    if rgb and not lad:
        binned_track.loc[mask, 'rgb'] = '248,248,248'
    else:
        if not lad:
            binned_track.loc[mask, 'rgb'] = '#F8F8F8'
        
    return binned_track
```

```python
chromsizes = bf.fetch_chromsizes('hg38')
chrombins = bf.binnify(chromsizes[:22], binsize=1000)
```

```python
lad_binned_df = bin_track(chrombins, lad_df, rgb=True, lad=True)
lad_binned_df = lad_binned_df.drop_duplicates()
lad_binned_df.head()
```

```python
hmmp_binned_df = bin_track(chrombins, hmmp_df, rgb=True)
hmmp_binned_df.head()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
hmmp_names = sorted(hmmp_binned_df.name.unique())
lad_names = sorted(lad_binned_df.name.unique())
```

```python
def comparison_matrix(track1, track2, track1_names, track2_names):
    labels = track1_names + track2_names
    M = confusion_matrix(track1.name, track2.name, labels=labels)
    M1 = M[:len(track1_names)-1, len(track1_names):]
    return M1
```

```python
plt.figure(figsize=(8, 5))
M = comparison_matrix(hmmp_binned_df, lad_binned_df, hmmp_names, lad_names) + 1
M_expected = np.outer(M.sum(axis=1), M.sum(axis=0))/M.sum()
sns.heatmap(np.log2(M/M_expected), annot=True,
           yticklabels=hmmp_names[:-1],
           xticklabels=lad_names, cmap='bwr', cbar=True, vmin=-1, vmax=1)
annotation1 = 'HMM3P'
annotation2 = 'LADS'
plt.xlabel(annotation2, fontsize=20)
plt.ylabel(annotation1, fontsize=20)
# if saveFigs:
#     plt.savefig(figure_path+'hiccomparison_OE_'+cell_line+'_'+assembly+'_'+annotation1+'_'+annotation2+'.svg',
#                 bbox_inches="tight")
```
