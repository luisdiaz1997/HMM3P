```python
import pandas as pd
import numpy as np
```

```python
gm_df = pd.read_csv('../annotations/GM12878_hg38_50kb.tsv', sep = '\t')
im_df = pd.read_csv('../annotations/IMR90_hg38_50kb.tsv', sep = '\t')
```

```python
mbb_to_b = (gm_df['HMM3P'] != im_df['HMM3P']) & (gm_df['HMM3P'] == 0) & (im_df['HMM3P'] == 3)
```

```python
gm_df[(gm_df.chrom=='chr2') & mbb_to_b]
```

```python
np.diff(gm_df[mbb_to_b]['start'])
```

```python
from astropy.convolution import convolve
```

```python
p1= convolve(gm_df[mbb_to_b]['start'].values, np.array([1, -2, 1]), normalize_kernel=False, boundary='fill')
p2 = np.where(p1==0, 0, 1)
```

```python
p3 = np.where(p2>0)[0]
```

```python
max_start = np.argmax(np.diff(p3))
max_end = max_start + 1
```

```python
gm_df[mbb_to_b].iloc[p3[max_start]:p3[max_end+1]]
```
